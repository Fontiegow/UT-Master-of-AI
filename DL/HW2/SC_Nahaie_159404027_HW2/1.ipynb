### 1. Mean Absolute Error (MAE)
$$MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$$

### 2. Mean Squared Error (MSE)
$$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

### 3. Root Mean Squared Error (RMSE)
$$RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$$

### 4. Coefficient of Determination (RÂ²)
$$R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}$$

### 5. Cross-Entropy Loss
$$L = - \sum_{i=1}^{n} \sum_{c=1}^{M} y_{i,c} \log(\hat{y}_{i,c})$$

```python
import numpy as np

y = np.array([3, 5, 2])
y_hat = np.array([2.5, 5.2, 1.7])

MAE = np.mean(np.abs(y - y_hat))
MSE = np.mean((y - y_hat)**2)
RMSE = np.sqrt(MSE)

print("MAE:", MAE)
print("MSE:", MSE)
print("RMSE:", RMSE)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate errors ranging from -3 to 3
errors = np.linspace(-3, 3, 100)

# Calculate Loss functions
mse_loss = errors ** 2
mae_loss = np.abs(errors)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(errors, mse_loss, label='MSE (Quadratic)', color='blue', linewidth=2)
plt.plot(errors, mae_loss, label='MAE (Linear)', color='red', linewidth=2, linestyle='--')

plt.title('Comparison of MAE and MSE Loss Functions', fontsize=14)
plt.xlabel('Error Value (y - y_hat)', fontsize=12)
plt.ylabel('Loss Value', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Adding annotations for differentiability
plt.annotate('Non-differentiable at 0', xy=(0, 0), xytext=(0.5, 2),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```
